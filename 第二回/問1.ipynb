```python
def make_X(mu, sigme, size):
    X = np.random.multivariate_normal(mu, sigma, size)
    return X
```

```python
def gaussian(X, mu, sigma):
    g = 1/np.sqrt((2*np.pi)**len(X.T))*np.sqrt(np.abs(sigma))
    print((X-mu).shape)
    gg = np.exp(-0.5 * (X-mu).dot(1/sigma).dot((X-mu).T))
    Z = g*gg
    return Z
```

```python
def zscore(x):
    xmean = x.mean(axis=0)
    xstd = np.std(x,axis=0)
    return (x - xmean) / xstd
```

```python
mu = np.array([0, 0])
sigma = np.array([[30, 20], [20, 50]])
size = 10000

X = zscore(make_X(mu,sigma, size))

sns.jointplot(X[:,0], X[:,1], xlim=(-20,20), ylim=(-20,20))
    
mu_x = X.mean(axis=0)
print("mu=",mu_x)

sigma_x = np.cov(X.T)
print("sigma=",sigma_x)
```
